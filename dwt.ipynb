```python
import pywt
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from  matplotlib.pyplot import figure

mydata = pd.read_csv("FinalData.csv")
# mydata = mydata.T
lst = mydata.iloc[0].to_numpy()
lst = lst[1:]
mydata = mydata.T
ind = mydata.index.values.tolist()
ind = ind[1:]
# print (ind)
# print (lst)

(cA, cD) = pywt.dwt(lst, 'db2', 'smooth')
A = pywt.idwt(cA, None, 'db2', 'smooth')
D = pywt.idwt(None, cD, 'db2', 'smooth')
# lst_rec = A + D #reconstruction of lst i.e data before decomposition

d = {'data': lst, 'linear components': A, 'non-linear components': D}
df = pd.DataFrame(d)
# df.index = ind

print (df)

# figure(num=None, figsize=(100, 10), dpi=80, facecolor='w', edgecolor='k')

df.plot(figsize=(100, 10), grid=True)
plt.show()
```
